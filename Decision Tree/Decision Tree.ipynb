```
#import librari
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```
#load data
bitfinexData = pd.read_csv("histdataBitfinex.csv")
```

```
#change timestamp
bitfinexData = bitfinexData.set_index(pd.DatetimeIndex(bitfinexData["Date"].values))
bitfinexData.index.name = "Date"
print(bitfinexData)
```

```
#neue Spalte ob preis hoch oder runter ist
bitfinexData["PriceUp"] = np.where(bitfinexData["Close"].shift(-1) > bitfinexData["Close"],1,0)
bitfinexData = bitfinexData.drop(columns = ["Date"])
bitfinexData.head(15)
```

```
#split the data set into a feature and a target data set
X = bitfinexData.iloc[:,0:bitfinexData.shape[1]-1].values
Y = bitfinexData.iloc[:,bitfinexData.shape[1]-1].values
```

```
#
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```
#Create and train the model (DecisionTreeClassifier)
tree = DecisionTreeClassifier().fit(X_train,Y_train)

#show how well the model did on the 

print(tree.score(X_test,Y_test))
```

```
#show 
tree_pred = tree.predict(X_test)
print(tree_pred)
```
